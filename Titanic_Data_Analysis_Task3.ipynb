```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic-dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
bins = [0, 12, 19, 59, 100]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df.groupby('AgeGroup')['Survived'].mean()
```

```python
df.groupby('Embarked')['Survived'].mean()
```

```python
df.groupby('FamilySize')['Survived'].mean()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

```

```python
corr = numeric_df.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby('FamilySize')['Survived'].mean().plot(kind='bar')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Family Size')
plt.show()
```

# Conclusion

- Children had higher survival rates.
- Certain embarkation ports showed better survival.
- Small families survived more than solo or large families.
- Feature engineering improved analysis quality.
